# MinMax with For/While Loops

```python
# N1: n ta son orasidan eng katta va eng kichigini topish

n = int(input("n = "))

maxi = 0
mini = 0
first = True
while n > 0:
    a = int(input("a = "))
    if first:
        mini = a
        maxi = a
        first = False
    if a > maxi:
        maxi = a
    if a < mini:
        mini = a
    n -= 1
print("eng kichigi:", mini, "eng kattasi:", maxi)
```

```python
### N2: n ta to'rtbuchak berilga. Eng kichik yuzaga ega to'rtburchak chiqarilsin

n = int(input("n = "))

first = True
yuza = 0
mini_a = 0
mini_b = 0

while n > 0:
    a = int(input("a = "))
    b = int(input("b = "))
    
    # birinchi kiritlgan qiymatni o'zlashtiradi
    if first:
        yuza = a * b
        mini_a = a
        mini_b = b
        first = False
        
    # ikkinchi qiymatdan boshlab, 
    # barcha a va b qiymatlarni kichikligini tekshiradi
    if yuza > a * b:
        yuza = a * b
        mini_a = a
        mini_b = b
    n -= 1
    
print("a:", mini_a, "| b:", mini_b, "| yuza:", yuza)
```

```python
# N3: n ta to'rtbuchak berilga. Eng katta perimeter ega to'rtburchak chiqarilsin

n = int(input("n = "))

first = True
per = 0
max_a = 0
max_b = 0

while n > 0:
    a = int(input("a = "))
    b = int(input("b = "))
    
    # birinchi kiritlgan qiymatni o'zlashtiradi
    if first:
        per = 2*a + 2*b
        max_a = a
        max_b = b
        first = False
        
    # ikkinchi qiymatdan boshlab, 
    # barcha a va b qiymatlarni kattalikka tekshiradi
    if per < 2*a + 2*b:
        per = 2*a + 2*b
        max_a = a
        max_b = b
    n -= 1
    
print("a:", max_a, "| b:", max_b, "| perimetr:", per)
```

```python
# N4: n ta son berilgan. Eng kichik element o'rnini aniqlash

n = int(input("n = "))

first = True
i = 1
mini = 0

while i < n + 1:
    
    # input kiritish
    a = int(input("a = "))
    
    # dastlabki qiymatni o'zlashtirish
    if first:
        son = i
        mini = a
        first = False
        
    # har bir qiymatni kichigligini tekshirish
    if mini > a:
        son = i
        mini = a
        
    # raqamlarni tartib raqamini sanab borish
    i += 1

print(son)
```

```python
# N5: n ta jismni hajmi va og'irligi berilgan. Zichligi eng kattasini topish

n = int(input("n = "))

first = True
zichlik = 0
max_m = 0
max_v = 0

while n > 0:
    m = int(input("m = "))
    v = int(input("v = "))
    
    # birinchi kiritlgan qiymatni o'zlashtiradi
    if first:
        zichlik = m / v
        max_m = m
        max_v = v
        first = False
        
    # ikkinchi qiymatdan boshlab, 
    # barcha a va b qiymatlarni kattalikka tekshiradi
    if zichlik < m / v:
        zichlik = m / v
        max_m = m
        max_v = v
    n -= 1
    
print("massa:", max_m, "| hajm:", max_v, "| zichlik:", zichlik)
```

```python
# N6: n ta son berilgan. Birinchi uchraga eng kichik va 
# oxirgi uchragan eng katta sonlar tartib raqami choqarilsin

n = int(input("n = "))

first = True
i = 1

while i < n + 1:
    a = int(input("a = "))
    
    # birinchi qiymatni o'zlashtir
    if first:
        min_bir = a
        max_oxir = a
        mb = i
        mo = i
        first = False
        
    # birinchi uchragan eng kichik raqam o'rni
    if min_bir > a:
        min_bir = a
        mb = i
        
    # oxirgi uchragan eng katta raqam o'rni
    if max_oxir <= a:
        max_oxir = a
        mo = i
    
    i += 1
    
print("Birinchi kichik:", mb, "| Oxirgi katta:", mo)
```

```python
# N13: n ta sonlar orasidan 
# birinchi uchragan eng katta toq son tartib raqami chiqarilsin

n = int(input("n = "))

i = 1
first = True
max_toq = 0

while i < n+1:
    a = int(input("a = "))
    
    # birinchi qiymatni o'zlashtir
    if first and a % 2 == 1:
        max_toq = a
        son = i
        first = False
        
    # son toq ekanligini va max_toq dan kattaligini tekshir
    if a % 2 == 1 and max_toq < a:
        max_toq = a
        son = i
    
    # sonlarni tartibini kuzatib bor
    i += 1

# toq son bolmasa, 0 print qil
if max_toq == 0:
    print(son-1)
else:
    print(son)
```

```python
# N22: n ta sonlar orasidan eng kichik ikkita qiymatni aniqlang.

n = int(input("n = "))

first = True
second = True
mini1 = 0
mini2 = 0
uch = 0

while n > 0:
    a = int(input("a = "))
    
    if first:
        mini1 = a
        first = False
        continue
        
    if not first and second:
        mini2 = a
        
        if mini2 < mini1:
            mini1, mini2 = mini2, mini1
        
        second = False
        continue
        
    if mini1 >= a:
        mini2 = mini1
        mini1 = a
    
    n -= 1
print(mini1, mini2)
```

```python
# N24: n ta sonlar berilgan. 
# Ikkita qo'shni sonlarni yig'indisi eng katta bo'lgan qiymat aniqlansin.

n = int(input("n = "))

# to'plamdagi birinchi son
one = int(input("a = "))
n -= 1
two = 0

# max qiymatga ega qo'shni sonlarni saqlash uchun
one_save = 0
two_save = 0

sum_of_two = 0
max_sum = 0
second = True

while n > 0:
    a = int(input("a = "))
    
    if second:
        two = a
        max_sum = one + two
        
        second = False
        n -= 1
        continue
    
    # oxirgi ikkita sonni yig'indisi
    sum_of_two = two + a
    
    # oxirgi ikkita sonlar yig'indisi, max yig'indidan katta bo'lib katsa,
    # ikkala sonlarni max sifatida saqlab qo'y, va yig'indini ham saqlab qo'y
    # aks holda, oxirgi ikki raqamni oddiy saqlab qo'y
    if sum_of_two > max_sum:
        one, two = two, a
        max_sum = sum_of_two
        one_save = one
        two_save = two
    else:
        one, two = two, a
    
    n -= 1
    
print("Birinchi:", one_save, "| Ikkinchi:", two_save, "| Yig'indi:", max_sum)
```

```python
# N25 Version 1: n ta sondan iborat to'plam berilgan.
# Ko'paytmasi eng kichik bo'lgan qo'shni sonlar index ini aniqlang.

n = int(input("n = "))

# to'plamning birinchi soni
one = int(input("a = "))
i = 2
two = 1

# ko'paytmasi kichik bo'lgan index larni saqlab boramiz
index_1 = 0
index_2 = 0

# oxirgi ikkita sonni ko'paytmasi
pro_of_two = one * two

# minimum ko'paytma qiymati
min_pro = pro_of_two

second = True

while i < n:
    a = int(input("a = "))
    
    # dastlabki ikki qiymatlar ko'paytmalari 
    # va indexlarini saqlash
    if second:
        two = a
        pro_of_two = one * two
        min_pro = pro_of_two
        
        # birnchi va ikkinchi indexlar
        index_1 = i - 1
        index_2 = i 
        
        second = False 
        continue
    
    # oxirgi ikki qo'shni sonlarni ko'paytmalari
    pro_of_two = two * a
      
    # agar oxirgi qo'shni sonlar ko'paytmasi minimum ko'paytmadan kichik bolsa,
    # uni min_pro ga saqla, va indexlarni o'zgartir
    # aks holda, keyingi qo'shni sonlarga o'tib ket
    if pro_of_two < min_pro:
        min_pro = pro_of_two
        
        index_1 = i 
        index_2 = i + 1
    else:
        one, two = two, a
    
    i += 1
    
print(index_1, index_2)
```
